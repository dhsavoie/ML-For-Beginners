```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection
```

```python
# load dataset
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape, f"-This dataset has {X.shape[0]} samples and {X.shape[1]} features") # this line was autofilled by copilot- cool!
print(X[0])
```

```python
X_2 = X[:, np.newaxis, 2] 
# q: what does np.newaxis do?
# a: https://stackoverflow.com/questions/29241056/how-does-numpy-newaxis-work-and-when-to-use-it 
# it looks like it adds a dimension to the array, it seems like it turned X into just it's third column (index 2)
```

```python
print(X_2.shape)
print(X_2[0])
# X is now just its third column, which is bmi (body mass index)
```

```python
# split the data into training/testing sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_2, y, test_size=0.33)
# this creates a 1/3 test set and a 2/3 training set
```

```python
print(X_train.shape[0], f"should be 2/3 of the original X, which is {442*0.67}")
```

```python
# now we can load and train the model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# now create a prediction using test data with the predict() function
y_pred = model.predict(X_test)
```

```python
# now we can plot some of the data to see how well the model did
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title("A Graph plot showing Diabetes Progression against BMI")
plt.show()
# this plot can predict the progression of diabetes based on the BMI of the patient
```
